```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
import re
import string
import matplotlib.pyplot as plt
```

```python
import chardet

rawdata = open("final_dataset.csv", "rb").read()
result = chardet.detect(rawdata)
encoding = result['encoding']
```

```python
data = pd.read_csv("final_dataset.csv", encoding=encoding)
```

```python
data.head(10)
```

```python
wanted_columns = ["review_bold", "ratings", "review", "verified", "date", "by", "profile_id", "most_rev", "helpful", "product", "review_sentiment"]

# Remove unwanted columns
data = data[wanted_columns]
```

```python
# Replace null values with empty strings in 'review_bold' and 'review' columns
data['review_bold'].fillna('', inplace=True)
data['review'].fillna('', inplace=True)

# Drop rows with null values in other columns
data.dropna(subset=["review_bold", "ratings", "review", "verified", "date", "by", "profile_id", "most_rev", "helpful", "product", "review_sentiment"], inplace=True)

# Reset index after dropping rows
data.reset_index(drop=True, inplace=True)

# Display the updated DataFrame
print(data.head())
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize



# Remove punctuation, convert text to lowercase, and remove stopwords
data['review_bold'] = data['review_bold'].apply(lambda x: ''.join([char for char in x if char not in string.punctuation]))
data['review_bold'] = data['review_bold'].apply(lambda x: x.lower())
data['review_bold'] = data['review_bold'].apply(lambda x: ' '.join([word for word in word_tokenize(x) if word not in set(stopwords.words('english'))]))

# Reset index after dropping rows
data.reset_index(drop=True, inplace=True)

# Display the updated DataFrame
print(data.head())
```

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(data['profile_id'],data['verified'], test_size=0.25, random_state=42)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Vectorize the text data
vectorizer = TfidfVectorizer()
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report


# Train the KNN model
knn = KNeighborsClassifier()
knn.fit(X_train_vec, y_train)
```

```python
# Predict on the test set
y_pred = knn.predict(X_test_vec)
```

```python
# Evaluate the model
print(classification_report(y_test, y_pred))
```

```python
from sklearn.svm import SVC

# Train the SVM model
svm = SVC()
svm.fit(X_train_vec, y_train)

# Predict on the test set
y_pred_svm = svm.predict(X_test_vec)

# Evaluate the SVM model
print(classification_report(y_test, y_pred_svm))
```

```python
from sklearn.linear_model import LogisticRegression

# Train the Logistic Regression model
log_reg = LogisticRegression()
log_reg.fit(X_train_vec, y_train)

# Predict on the test set
y_pred_log_reg = log_reg.predict(X_test_vec)

# Evaluate the Logistic Regression model
print(classification_report(y_test, y_pred_log_reg))
```

```python
from sklearn.tree import DecisionTreeClassifier

# Train the Decision Tree model
dt = DecisionTreeClassifier()
dt.fit(X_train_vec, y_train)

# Predict on the test set
y_pred_dt = dt.predict(X_test_vec)

# Evaluate the Decision Tree model
print(classification_report(y_test, y_pred_dt))
```

```python
from sklearn.naive_bayes import MultinomialNB

# Train the Naive Bayes model
nb = MultinomialNB()
nb.fit(X_train_vec, y_train)

# Predict on the test set
y_pred_nb = nb.predict(X_test_vec)

# Evaluate the Naive Bayes model
print(classification_report(y_test, y_pred_nb))
```

```python
from sklearn.ensemble import RandomForestClassifier

# Train the Random Forest model
rf = RandomForestClassifier()
rf.fit(X_train_vec, y_train)

# Predict on the test set
y_pred_rf = rf.predict(X_test_vec)

# Evaluate the Random Forest model
print(classification_report(y_test, y_pred_rf))
```

```python
from sklearn.metrics import accuracy_score

# Assuming 'y_test' is the true labels and 'X_test_vec' is the test data for each model
# Replace 'X_test_vec' with your actual test data after vectorization

# KNN
knn_accuracy = accuracy_score(y_test, knn.predict(X_test_vec))
print("KNN Accuracy:", knn_accuracy)

# Logistic Regression
lr_accuracy = accuracy_score(y_test, log_reg.predict(X_test_vec))
print("Logistic Regression Accuracy:", lr_accuracy)

# Support Vector Machine
svm_accuracy = accuracy_score(y_test, svm.predict(X_test_vec))
print("SVM Accuracy:", svm_accuracy)

# Naive Bayes
nb_accuracy = accuracy_score(y_test, nb.predict(X_test_vec))
print("Naive Bayes Accuracy:", nb_accuracy)

# Decision Tree Classifier
dt_accuracy = accuracy_score(y_test, dt.predict(X_test_vec))
print("Decision Tree Accuracy:", dt_accuracy)

# Random Forest Classifier
rf_accuracy = accuracy_score(y_test, rf.predict(X_test_vec))
print("Random Forest Accuracy:", rf_accuracy)
```
